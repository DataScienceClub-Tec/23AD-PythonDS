```python
import requests # this lib is used to get the html from the website
from bs4 import BeautifulSoup # this lib is used to parse the html
```

```python
## This is a dictionary
headers = {
    "User-Agent:":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"
} # User agent is specific to the browser you are using

url = "https://finance.yahoo.com/quote/TSLA" 
r = requests.get(url) # this is the html of the website
soup = BeautifulSoup(r.text, "html.parser") # this parses the html of the websiter
print(r.text)
price = soup.find_all() # this is the price of the stock
print(price)
```
